```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
import sqlite3 as sql
```

```python
url = 'https://web.archive.org/web/20230902185655/https://en.everybodywiki.com/100_Most_Highly-Ranked_Films'
```

```python
html_page=requests.get(url)
data=bs(html_page.text, "html.parser")
table=data.find_all('tbody')
rows=table[0].find_all("tr")
```

```python
df = pd.DataFrame(columns=["Average Rank","Film","Year"])

count=0
for n in rows:
    if count<51:
        if count==0:
            pass
        else:
            cols=n.findAll("td")
            dict={"Average Rank":cols[0].contents[0],
                  "Film":cols[1].contents[0],
                  "Year":cols[2].contents[0]}
            df_temp=pd.DataFrame(dict, index=[0])
            df=pd.concat([df,df_temp], ignore_index=True)
            # for n in cols:
                # print(n)
                # for k in n:
                    # print(k)
            # print(cols[0:3])
        count+=1
```

```python
#Salvataggio in cs
df.to_csv("top_film.csv", index=False)
```

```python
#Salvataggio in database
conn=sql.connect("Movies.db")
df.to_sql("Top_50",conn,if_exists='replace', index=False)
conn.close()
```
